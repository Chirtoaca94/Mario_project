```


df_4 = _deepnote_execute_sql("""ALTER TABLE MORE_INFO ADD COLUMN CHARACTER_ID INT
REFERENCES CHARACTERS(CHARACTER_ID);

""", 'SQL_1C71274F_4367_4FAF_A3BC_39033D0FA0D6')
df_4
```

```


df_6 = _deepnote_execute_sql("""SELECT * 
FROM CHARACTERS 
JOIN
MORE_INFO 
ON CHARACTERS.character_id = MORE_INFO.CHARACTER_ID;
""", 'SQL_1C71274F_4367_4FAF_A3BC_39033D0FA0D6')
df_6
```
